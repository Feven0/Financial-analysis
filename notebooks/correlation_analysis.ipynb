# Task 3: Correlation between News Sentiment and Stock Movement

This notebook analyzes the relationship between daily news sentiment scores and stock price returns.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys

# Add src to path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from src.data_loader import load_news_data, load_stock_data
from src.sentiment import apply_sentiment_analysis, aggregate_daily_sentiment
from src.correlation import calculate_stock_returns, merge_news_stock, calculate_correlation

sns.set(style="whitegrid")
```

```python
# Load data
news_path = '../data/raw/news_data/raw_analyst_ratings.csv'
stock_folder = '../data/raw/yfinance_data/Data'

# Sampling news for speed (optional)
news_df = pd.read_csv(news_path)
aapl_df = load_stock_data(stock_folder, 'AAPL')
```

## 1. Sentiment Analysis

Calculate sentiment scores for the headlines.

```python
# Filter news for AAPL
aapl_news = news_df[news_df['stock'] == 'AAPL'].copy()
aapl_news = apply_sentiment_analysis(aapl_news)
daily_sentiment = aggregate_daily_sentiment(aapl_news)
daily_sentiment.head()
```

## 2. Stock Returns

Calculate daily returns for AAPL.

```python
aapl_df = calculate_stock_returns(aapl_df)
aapl_df.head()
```

## 3. Correlation Analysis

Merge the datasets and calculate correlation.

```python
merged_df = merge_news_stock(daily_sentiment, aapl_df)
correlation = calculate_correlation(merged_df)
print(f"Correlation between daily sentiment and AAPL returns: {correlation:.4f}")
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='sentiment', y='returns', data=merged_df)
plt.title('Correlation: News Sentiment vs Daily Returns (AAPL)')
plt.show()
```
