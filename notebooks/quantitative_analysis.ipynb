# Task 2: Quantitative Analysis using TA-Lib and PyNance

This notebook focuses on technical indicators and financial metrics for stock data.

```python
import pandas as pd
import talib
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys

# Add src to path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from src.data_loader import load_stock_data
from src.indicators import calculate_moving_averages, calculate_rsi, calculate_macd

sns.set(style="whitegrid")
```

```python
# Load AAPL data
stock_folder = '../data/raw/yfinance_data/Data'
aapl_df = load_stock_data(stock_folder, 'AAPL')
aapl_df['Date'] = pd.to_datetime(aapl_df['Date'])
aapl_df.set_index('Date', inplace=True)
aapl_df.head()
```

## 1. Technical Indicators (TA-Lib)

### Moving Averages

```python
aapl_df = calculate_moving_averages(aapl_df)
aapl_df[['Close', 'SMA_20', 'SMA_50']].tail(100).plot(figsize=(12, 6))
plt.title('AAPL Price and Moving Averages')
plt.show()
```

### Relative Strength Index (RSI)

```python
aapl_df = calculate_rsi(aapl_df)
plt.figure(figsize=(12, 4))
aapl_df['RSI'].tail(100).plot()
plt.axhline(70, color='r', linestyle='--')
plt.axhline(30, color='g', linestyle='--')
plt.title('AAPL RSI (14)')
plt.show()
```

### MACD

```python
aapl_df = calculate_macd(aapl_df)
aapl_df[['MACD', 'MACD_signal']].tail(100).plot(figsize=(12, 4))
plt.title('AAPL MACD')
plt.show()
```
